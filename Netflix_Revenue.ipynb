```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings

filterwarnings("ignore")
%matplotlib inline
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/oishbdh/Datasets/main/archive/Netflix%20Revenue.csv")
data
```

```python
#Remove commas and convert 'Netflix Global Users' to numeric

#data['Netflix Global Users'] = data['Netflix Global Users'].str.replace(',', '').astype(float)

#Let's now extract the Year from the Date column

data['Year'] = pd.to_datetime(data['Date']).dt.year
```

```python
data.describe()
```

```python
#Convert the 'Date' to 'Datetime' format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
#Let's sort the data by date

data = data.sort_values('Date')

#Lets plot the revenue trends over time

plt.figure(figsize = (12, 6))
plt.plot(data['Date'], data['Global Revenue'])
plt.title("Netflix Global Revenue Over Time")
plt.xlabel('Date')
plt.ylabel('Global Revenue')
plt.xticks(rotation = 45)

#Let's modify the y axis tick label

plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.2f}".format(x)))

plt.show
```

```python
#Group the data by year and calculate the average users
average_users = data.groupby('Year')['Netflix Global Users'].mean()

#Time to plot average global users over time
plt.figure(figsize = (12, 6))
plt.plot(average_users.index, average_users.values)
plt.title('Average Netflix Global Users By Year')
plt.xlabel('Year')
plt.ylabel('Average Global Users')
plt.xticks(rotation = 45)

#modify the yaxis tick label

plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.2f}".format(x)))
plt.show()
```

```python
#Time to sort the data by date
data.sort_values('Date', inplace = True)

#Plotting global users trend over time
plt.figure(figsize = (12, 6))
plt.plot(data['Date'], data['Netflix Global Users'], color = 'Red')
plt.title('Netflix Global Users Trend', fontsize = 16)
plt.xlabel('Date', fontsize = 13)
plt.ylabel('Global Users', fontsize = 13)
plt.xticks(rotation = 45)
plt.grid(True)

#modify yaxis tick label
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.2f}".format(x)))
plt.show()
```

```python
#plotting revenue by region
#first we drop rows with missing values in the revenue columns

data_clean = data.dropna(subset = ['UACN Revenue', 'EMEA Revenue', 'LATM Revenue', 'APAC Revenue'])

#Lets calculate the sum of our revenue by region

revenue_by_region = data_clean[['UACN Revenue', 'EMEA Revenue', 'LATM Revenue', 'APAC Revenue']].sum()

#We can go ahead to plot a bar chart of revenue by region
plt.figure(figsize = (10, 6))
revenue_by_region.plot(kind = 'bar', color = 'red')
plt.title('Revenue by region', fontsize = 16)
plt.xlabel('Region', fontsize = 12)
plt.ylabel('Revenue', fontsize = 12)
plt.xticks(rotation = 8)
plt.grid(axis = 'y')

#modify the y tick label
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.2f}".format(x)))
plt.show
```
